```python
available_labels_sided = [labels_sided[i-1] for i in np.unique(localAdjusted_volume) if i > 0]
available_labels_unsided = set([labelMap_sidedToUnsided[name] for name in available_labels_sided ])
```

```python
contours_allSecs = extract_contours_from_labeled_volume(stack, volume='globalAligned',
                                                          downsample_factor=16,
                                                          labels=['BackG'] + labels_sided,
                                                          force=True,
                                                          extrapolate_num_section=0)
```

```python
# username = 'localAdjusted'
username = 'globalAligned'
```

```python
for sec, contours in contours_allSecs.iteritems():
    
    print sec
    
    autoAnnotations_dir = create_if_not_exists(os.environ['LABELING_DIR'] + '/' + stack + '/' + '%04d'%sec)
                
    valid_labels = set(available_labels_sided) & set(contours.keys())
    print valid_labels
    
    if len(valid_labels) == 0:
        sys.stderr.write('No valid labels exist.\n')
        continue

    new_annotations = []
    
    for name_s in valid_labels:
                
        cnt = contours[name_s]

        if len(cnt) < 3:
            sys.stderr.write('initial contour has less than 3 vertices. \n')
            continue
        
        new_lm = {}
        new_lm['label'] = labelMap_sidedToUnsided[name_s]
        new_lm['vertices'] = cnt.astype(np.int) * 8
        new_lm['labelPos'] = new_lm['vertices'].mean(axis=0)
        new_lm['refVertices'] = np.array(cnt).copy() * 8
        new_lm['subtype'] = PolygonType.CLOSED

        new_annotations.append(new_lm)
        
        
    timestamp = datetime.datetime.now().strftime("%m%d%Y%H%M%S")

    autoAnnotation_filepath = autoAnnotations_dir + '/%(stack)s_%(sec)04d_%(username)s_%(timestamp)s_consolidated.pkl' % \
                        {'stack': stack, 'sec': sec, 'timestamp': timestamp, 'username': username}

    pickle.dump(new_annotations, open(autoAnnotation_filepath, 'w'))
```
