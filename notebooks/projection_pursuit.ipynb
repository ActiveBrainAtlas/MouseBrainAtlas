```python
def centering_and_whitening(X):
    
    # centering
    
    mean = X.mean(axis=1)
    X = X - mean
    
    # whitening
    XXT = np.dot(X, X.T)
    D, U = np.linalg.eigh(XXT)
    D_negsqrt = np.diagflat(1./np.sqrt(D))
    whitening_matrix = np.dot(D_negsqrt, U.T)
    X = np.dot(whitening_matrix, X)
    
    # test the effect of centering and whitening
    # print np.linalg.norm(np.dot(X, X.T) - eye(X.shape[0]))
    # plt.scatter(X[2,:], X[8,:]);
    
    return X, mean, whitening_matrix
```

```python
import cv2

def highlight_patches(target_patch_indices, template_patch_index):
    
    img_vis = img.copy() * mask2[..., np.newaxis]
    
    # draw a blue box around the template patch
    (ymin, ymax, xmin, xmax) = patch_positions[template_patch_index]
    cv2.rectangle(img_vis, (xmin, ymin), (xmax, ymax), (0,0,255), 2)

    # draw a red box around each target patch
    for (ymin, ymax, xmin, xmax) in patch_positions[target_patch_indices]:
        cv2.rectangle(img_vis, (xmin, ymin), (xmax, ymax), (255,0,0), 2)
    
    return display(img_vis)
```

```python
from skimage.transform import rotate
from sklearn.preprocessing import normalize

def project_all_to_template(template_patch_index):
    '''
    require global variables `patch_positions` and `X`
    return two num_patches-by-1 arrays
    '''
    
    # retrieve template position
    (ymin, ymax, xmin, xmax) = patch_positions[template_patch_index]
    
    # enlarge template window
    cushion_ymin = ymin - 100
    cushion_xmin = xmin - 100
    cushion_ymax = ymax + 100
    cushion_xmax = xmax + 100
    
    # generate rotated versions of the template
    rotated_templates = []
    for angle in range(0, 360, 30):
        rotated_template = rotate(img_gray[cushion_ymin:cushion_ymax, cushion_xmin:cushion_xmax], 
                                  angle, resize=False)[100:-100, 100:-100]
        if rotated_template.size != 3600:
            raise Exception, 'enlarged window exceeds border'
        rotated_templates.append(rotated_template.flat)

    # d x num_angles
    rotated_templates = np.column_stack(rotated_templates)
    
    rotated_templates = np.dot(whitening_matrix, rotated_templates - mean[:, np.newaxis])
    rotated_templates = normalize(rotated_templates, norm='l2', axis=0)
    
    # project all patches to all rotated template versions
    projval_allrot = np.dot(rotated_templates.T, X)
    
    # use the largest response among all rotations
    projval_bestrot = projval_allrot.max(axis=0)
    projval_argbestrot = projval_allrot.argmax(axis=0)
    
    # ignore the template patch itself
    projval_bestrot[template_patch_index] = 0
    
    return projval_bestrot, projval_argbestrot
```

```python
def compute_projection_index(p):

    excess_kurtosis = 1./len(p)*((p-p.mean())**4).sum()/p.std()**4 - 3
#     print excess_kurtosis
    return excess_kurtosis
    
def detect_targets(p):
    
    # compute histogram of projected values
    hist, bins = np.histogram(p, bins=1000)
    
    # plot histogram
    if VERBOSE:
        plt.hist(p, bins);
    
    # find the first zero touching, and use it as threshold
    # patches to the right of threshold is regarded targets
    zero_touchings = np.where(hist==0)[0]
    first_zero_touching = zero_touchings[where(zero_touchings > np.argmax(hist))[0][0]]
    thresh = bins[first_zero_touching]
    
    if VERBOSE:
        print 'threshold', thresh
    
    target_patch_indices = where(p > thresh)[0]
    target_patch_indices_sorted = target_patch_indices[np.argsort(p[target_patch_indices])[::-1]]

    return target_patch_indices_sorted
```

```python
pis = []
template_indices = []
for template_patch_index in np.random.randint(0, len(patches), 1000):
    try:
        projval_bestrot, projval_argbestrot = project_all_to_template(template_patch_index)
        target_patch_indices = detect_targets(projval_bestrot)
        pi = compute_projection_index(projval_bestrot)
        pis.append(pi)
        template_indices.append(template_patch_index)
    except Exception, e:
        print e
        pass
```
