For each expansion cluster, vote for its border superpixels, which includes both surrounding neighbors (called _surrounds_) and member superpixels at the outermost layer (called _frontiers_). 
Currently every border superpixels gets 1 vote from its expansion cluster, but it might also make sense to make the voting weight vary based on the cluster size.

```python
sp_votes = np.zeros((n_superpixels,), dtype=np.float)

border_sps = []

for j, cluster in enumerate(cluster_sps):
    
    surrounds = set([i for i in set.union(*[neighbors[c] for c in cluster]) if i not in cluster and i != -1])
    frontiers = set.union(*[neighbors[c] for c in surrounds]) & set(cluster)

    border_sps.append(surrounds | frontiers)
    
#     weight = 1./np.sqrt(len(cluster))
    weight = 1.
    sp_votes[list(surrounds)] += weight
```
